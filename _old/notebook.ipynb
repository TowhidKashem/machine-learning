### Different ways to load an image in python

```python
file = './teachable-machine/dataset/cats/abyssinian-2.jpeg'
```

#### IPython:

```python
from IPython.display import Image # `pipenv install ipython`

Image(filename=file, width=224, height=224)
```

#### Tensorflow Keras:

```python
from tensorflow.keras.preprocessing import image # `pipenv install tensorflow`
from PIL import Image # `pipenv install pillow`

image.load_img(file, target_size= (224, 224))
```

#### Matlab:

```python
from matplotlib import pyplot # `pipenv install matplotlib`

# pyplot.imshow(file)
```

#### Opencv python (most common):

*this is like the graphicsmagick of the python ecosystem, can do a lot of different image processing with this lib*

```python
import cv2 # `pipenv install opencv-python`

img = cv2.imread(file) # load image
img = cv2.resize(img, (224, 224)) # resize
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # by default cv2 images are bgr so convert to rgb

pyplot.imshow(img) # display image using matplot
```

### Pillow:

```python
from PIL import Image # `pipenv install pillow`

img = Image.open(file)

img = img.resize((224, 224))

pyplot.imshow(img)
```
